## 파워 시리즈란?

$$
c_0 + c_1(x-a)^1 + c_2(x-a)^2 + ... +  + c_k(x-a)^k + ...
$$

## 테일러 시리즈란?

테일러 시리즈는 파워 시리즈의 상수를 아래와 같이 정의합니다.

$$
c_k = {f^{(k)}(a) \over k!}
$$

특히 $a=0$일 때 Maclaurin 시리즈라고 부릅니다.

## 테일러 시리즈 예제

아래 함수를 사용하여 실습합니다.

$$
f(x) = {1 \over x}
$$

$x = a$ 근처에서 위 함수의 2차 테일러 다항식은 다음과 같습니다.

$$
f(x) \approx f(a) - {(x-a) \over a^2} + {(x-a)^2 \over a^3}
$$

```python
def f(x):
    return 1/x

def f_taylor_polynomial_order_2(x, a):
    return f(a) - (x-a)/a**2 + (x-a)**2/a**3
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.2, 1.9, 0.1)
print(x)
a = 1

plt.plot(x, f(x), label='1/x')
plt.plot(x, f_taylor_polynomial_order_2(x, a), label='Order 2 approximation')
plt.xlim(0.0, 2.0)
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.arange(0.5, 5.6, 0.1)
a = 3

plt.plot(x, f(x), label='1/x')
plt.plot(x, f_taylor_polynomial_order_2(x, a), label='Order 2 approximation')
plt.xlim(0.1, 6.0)
plt.grid(True)
plt.legend()
plt.show()
```

## 더 많은 예제

$$
g(x) = e^x
$$

$$
g(x) = e^a + {e^a \over 1!}(x-a)^1 + {e^a \over 2!}(x-a)^2 + ... + {e^a \over k!}(x-a)^k + ...
$$

$x=0$ 근처에서 테일러 시리즈를 구하려면 $a=0$으로 지정하면 됩니다.

$$
g(x) = 1 + {x^1 \over 1!} + {x^2 \over 2!} + ... + {x^k \over k!} + ...
$$

$x=0$ 근처에서 k차 테일러 시리즈는 다음과 같습니다.

$$
g(x) \approx 1 + {x^1 \over 1!} + {x^2 \over 2!} + ... + {x^k \over k!}
$$

```python
import math

def g(x):
    return np.exp(x)
    
def g_taylor_polynomial(x, a, order=2):
    v = 1
    for i in range(1, order+1):
        v += x**i / math.factorial(i)
    return v
```

```python
x = np.arange(-3.0, 3.1, 0.1)
a = 0

plt.plot(x, g(x), label='Actual')
plt.plot(x, g_taylor_polynomial(x, a, 2), label='Order 2')
plt.plot(x, g_taylor_polynomial(x, a, 3), label='Order 3')
plt.plot(x, g_taylor_polynomial(x, a, 4), label='Order 4')
plt.plot(x, g_taylor_polynomial(x, a, 5), label='Order 5')
plt.xlim(-3.0, 3.0)
plt.grid(True)
plt.legend()
plt.show()
```
